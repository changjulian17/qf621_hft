```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("data/20250625/strategy_performance_detailed.csv")
```

```python
df.head(30)
```

```python
# Plot cumulative returns per strategy
returns = ['OBIVWAP_Returns', 'MeanRev_Returns', 'Portfolio_Returns']
df.set_index('ticker')[returns].plot(kind='bar', title='Strategy Returns by Ticker', figsize=(10, 5))
plt.ylabel("Return")
plt.tight_layout()
plt.show()
```

```python
# Plot Sharpe ratios
sharpes = ['OBIVWAP_Sharpe', 'MeanRev_Sharpe', 'Portfolio_Sharpe']
df.set_index('ticker')[sharpes].plot(kind='bar', title='Sharpe Ratios by Ticker', figsize=(10, 5))
plt.ylabel("Sharpe Ratio")
plt.tight_layout()
plt.show()
```

```python
# Plot number of trades
trades = ['OBIVWAP_Trades', 'MeanRev_Trades', 'Portfolio_Trades']
df.set_index('ticker')[trades].plot(kind='bar', title='Trade Counts by Strategy', figsize=(10, 5))
plt.ylabel("Trades")
plt.tight_layout()
plt.show()
```
